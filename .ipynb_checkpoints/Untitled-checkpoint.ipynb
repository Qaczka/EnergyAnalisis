```python
import pandas as pd
import numpy as np
```

```python
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
import sklearn
```

```python
%matplotlib inline
```

```python
energy = pd.read_excel('ENB2012_data.xlsx')
```

```python
energy.head()
```

```python
energy.info()
```

```python
energy.describe()
```

```python
sns.pairplot(energy)
```

```python
sns.distplot(energy['Y1'])
```

```python
energy.corr()
```

```python
sns.heatmap(energy.corr())
```

```python
energy.columns
```

```python
X = energy[['X1', 'X2', 'X3', 'X4', 'X5', 'X6', 'X7', 'X8']]
```

```python
y = energy['Y1']
```

```python
from sklearn.model_selection import train_test_split
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.25, random_state=101)
```

```python
from sklearn.linear_model import LinearRegression
```

```python
model = LinearRegression()
```

```python
model.fit(X_train, y_train)
```

```python
print(model.intercept_)
```

```python
model.coef_
```

```python
cdf = pd.DataFrame(model.coef_, X.columns, columns=['Coeff'])
```

```python
cdf
```

```python
predictions = model.predict(X_test)
```

```python
plt.scatter(y_test, predictions)
```

```python
sns.distplot((y_test-predictions))
```

```python
from sklearn import metrics
```

```python
metrics.mean_absolute_error(y_test, predictions)
```

```python
metrics.mean_squared_error(y_test, predictions)
```

```python
np.sqrt(metrics.mean_squared_error(y_test, predictions))
```

```python
model.score(X,y)
```

```python
metrics.r2_score(y_test, predictions)
```

```python
y2 = energy['Y2']
```

```python
X_train, X_test, y2_train, y2_test = train_test_split(X, y2, test_size=0.20, random_state=123)
```

```python
lineRegression_y2 = LinearRegression()
```

```python
print(lineRegression_y2.intercept_)
```

```python
lineRegression_y2.coef_
```

```python
cdf = pd.DataFrame(lineRegression_y2.coef_, X.columns, columns=['Coeff'])
```

```python
cdf
```

```python
pred_y2 = lineRegression_y2.predict(X_test)
```

```python
plt.scatter(y2_test, pred_y2)
```

```python
sns.distplot((y2_test-pred_y2))
```

```python
metrics.mean_absolute_error(y2_test, pred_y2)
```

```python
metrics.mean_squared_error(y2_test, pred_y2)
```

```python
np.sqrt(metrics.mean_squared_error(y2_test, pred_y2))
```
